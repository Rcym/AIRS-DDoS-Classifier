```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'C:\Users\RACYM\Desktop\M2_PFE\datasets\NSL-KDD\KDDTrain+.txt')
dataset
```

```python
dataset.columns = ['duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes', 'land', 'wrong_fragment', 'urgent', 'hot', 'num_failed_logins', 'logged_in', 'num_compromised', 'root_shell', 'su_attempted', 'num_root', 'num_file_creations', 'num_shells', 'num_access_files', 'num_outbound_cmds', 'is_host_login', 'is_guest_login', 'count', 'srv_count', 'serror_rate', 'srv_serror_rate', 'rerror_rate', 'srv_rerror_rate', 'same_srv_rate', 'diff_srv_rate', 'srv_diff_host_rate', 'dst_host_count', 'dst_host_srv_count', 'dst_host_same_srv_rate', 'dst_host_diff_srv_rate', 'dst_host_same_src_port_rate', 'dst_host_srv_diff_host_rate', 'dst_host_serror_rate', 'dst_host_srv_serror_rate', 'dst_host_rerror_rate', 'dst_host_srv_rerror_rate', 'attack', 'level']
dataset
```

```python
## plot the graph of the number of rows for each level
dataset.level.value_counts().plot(kind='bar')
plt.title('Number of rows for each level')
plt.xlabel('Level')
plt.ylabel('Number of rows')
plt.show()
```

```python
dataset.attack.value_counts().plot(kind='bar')
plt.title('Number of rows for each attack')
plt.xlabel('Attack type')
plt.ylabel('Number of rows')
plt.show()
```

```python
# plotting the number of rows for each attack type without normal and neptune
dataset.attack.value_counts().drop('normal').drop('neptune').plot(kind='bar')
plt.title('Number of rows for each attack')
plt.xlabel('Attack type')
plt.ylabel('Number of rows')
plt.show()
```
